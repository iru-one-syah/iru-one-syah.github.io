## Contoh 1

Diketahui:

<div align="center">
$x[n] = 4 \delta[n] + \delta[n-1] + 2 \delta[n-2] + 5 \delta[n-3]$
</div>

<br>

<div align="center">
dan
</div>

<br>

<div align="center">
$h[n] = \delta[n+1] + 2 \delta[n] - \delta[n-1]$
</div>

<br>

```octave
x_n = [4 1 2 5];
n_x = 0:3; % n = 0,1,2,3

h_n = [1 2 -1];
n_h = -1:1; % n = -1,0,1

subplot(211)
stem(n_x,x_n,'r','LineWidth',5)
ylim([min([x_n h_n])-1 max([x_n h_n])+1])
ylabel('x[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);

subplot(212)
stem(n_h,h_n,'r','LineWidth',5)
ylim([min([x_n h_n])-1 max([x_n h_n])+1])
ylabel('h[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);
```

Hitunglah:

$y[n] = x[n] * h[n]$

<br>

```octave
y_n = conv(x_n,h_n);
n_y = min(n_x)+min(n_h):max(n_x)+max(n_h);


figure(1,'position',[0 0 500 175])
stem(n_y,y_n,'r','LineWidth',5)
ylabel('y[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);
ylim([min(y_n)-1 max(y_n)+1])

disp(' n     y[n]')
disp('-----------')
for i = 1:length(y_n)
    printf(' %d \t %d \n',n_y(i),y_n(i));
end
```
