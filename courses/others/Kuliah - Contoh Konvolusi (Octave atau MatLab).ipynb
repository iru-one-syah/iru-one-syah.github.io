## Contoh 1

Diketahui:

<div align="center">
$x[n] = 4 \delta[n] + \delta[n-1] + 2 \delta[n-2] + 5 \delta[n-3]$
</div>

<br>

<div align="center">
dan
</div>

<br>

<div align="center">
$h[n] = \delta[n+1] + 2 \delta[n] - \delta[n-1]$
</div>

<br>

```octave
x_n = [4 1 2 5];
n_x = 0:3; % n = 0,1,2,3

h_n = [1 2 -1];
n_h = -1:1; % n = -1,0,1

subplot(211)
stem(n_x,x_n,'r','LineWidth',5)
ylim([min([x_n h_n])-1 max([x_n h_n])+1])
ylabel('x[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);

subplot(212)
stem(n_h,h_n,'r','LineWidth',5)
ylim([min([x_n h_n])-1 max([x_n h_n])+1])
ylabel('h[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);
```

Hitunglah:

$y[n] = x[n] * h[n]$

<br>

```octave
y_n = conv(x_n,h_n);
n_y = min(n_x)+min(n_h):max(n_x)+max(n_h);


figure(1,'position',[0 0 500 175])
stem(n_y,y_n,'r','LineWidth',5)
ylabel('y[n]')
xlabel('n')
xlim([-4 7])
set(gca, 'xtick',-4:7);
ylim([min(y_n)-1 max(y_n)+1])

disp(' n     y[n]')
disp('-----------')
for i = 1:length(y_n)
    printf(' %d \t %d \n',n_y(i),y_n(i));
end
```

## Contoh 2

Diketahui:

<div align="center">
    $x[n] = \alpha^n u[n]$ <br> <br>
    $h[n] = u[n]$
</div>

dimana $0<\alpha<1$

```octave
a = 0.9;

n_x = -60:60;
x_n = (a.^n_x).*(n_x>=0);

n_h = n_x;
h_n = (n_h>=0);

subplot(211)
stem(n_x,x_n,'r','LineWidth',5)
ylim([min([x_n h_n]) max([x_n h_n])])
ylabel('x[n]')
xlabel('n')
xlim([-20 20])

subplot(212)
stem(n_h,h_n,'r','LineWidth',5)
ylim([min([x_n h_n]) max([x_n h_n])])
ylabel('h[n]')
xlabel('n')
xlim([-20 20])
```

Hitunglah:

$y[n] = x[n] * h[n]$

<br>

```octave
y_n = conv(x_n,h_n);
n_y = min(n_x)+min(n_h):max(n_x)+max(n_h);


figure(1,'position',[0 0 525 200])
stem(n_y,y_n,'r','LineWidth',5)
ylabel('y[n]')
xlabel('n')
xlim([-20 20])
```

## Contoh 3

Diketahui:

<div align="center">
$ x[n] =
  \begin{cases}
    1       & \quad 0\leq n \leq 4\\
    0       & \quad \text{otherwise}
  \end{cases}
$
</div>

dan

<div align="center">
$ h[n] =
  \begin{cases}
    \alpha^n & \quad 0\leq n \leq 6\\
    0        & \quad \text{otherwise}
  \end{cases}
$
</div>

```octave
n_x = 0:4;
x_n = ones(size(n_x));

a = 1.5;
n_h = 0:6;
h_n = a.^n_h;

subplot(211)
stem(n_x,x_n,'r','LineWidth',5)
ylabel('x[n]')
xlabel('n')
xlim([-13 13])
set(gca, 'xtick',-13:13);

subplot(212)
stem(n_h,h_n,'r','LineWidth',5)
ylabel('h[n]')
xlabel('n')
xlim([-13 13])
set(gca, 'xtick',-13:13);
```

Hitunglah:

$y[n] = x[n] * h[n]$

<br>

```octave
y_n = conv(x_n,h_n);
n_y = min(n_x)+min(n_h):max(n_x)+max(n_h);


figure(1,'position',[0 0 525 200])
stem(n_y,y_n,'r','LineWidth',5)
ylabel('y[n]')
xlabel('n')
xlim([-13 13])
set(gca, 'xtick',-13:13);
```
